```python
import matplotlib.pyplot as plt
import numpy as np
import wave
import sys

%matplotlib inline
# открываем .wav файл
spf = wave.open('f.wav','r') 

signal = spf.readframes(-1)
signal = np.fromstring(signal, 'Int16')
fs = spf.getframerate()

Time=np.linspace(0, len(signal)/fs, num=len(signal))
# строим график
plt.figure(1)
plt.xlabel("Time")
plt.ylabel("Signal")
plt.title('Отрывок из аудиофайла')
plt.plot(Time,signal)
plt.axhline(y=max(signal), color='y') # жёлтая линия находится на уровне максимального значения сигнала
plt.show()
```
